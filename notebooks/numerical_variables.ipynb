```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/ai_job_dataset.csv")
data.head()
```

# About

The AI & Data Job Salaries and Skills Dataset 2024-2025 is synthetically prepared and provides a detailed, global overview of job listings in artificial intelligence, data science, machine learning, and related fields. This dataset aggregates job postings from October 2024 to July 2025, capturing a wide spectrum of roles, required skills, compensation, and organizational attributes across multiple countries and industries.

With over 15,000 entries, each record details the job title, salary (in USD and local currency), experience level, employment type, company location and size, remote work ratio, required technical skills, education and experience requirements, industry, posting and application dates, job description length, benefits score, and company name. This rich dataset is designed for data scientists, career analysts, HR professionals, and researchers interested in AI workforce trends, salary benchmarking, and skills demand analysis.

```python
def distribution(colum_name):
    fig,axes =plt.subplots(2,2,figsize=(15,15))
    sns.kdeplot(data=data[colum_name],ax=axes[0,0],fill=True,alpha=0.6,color="#2E86C1")

    sns.histplot(data=data[colum_name],ax=axes[0,1],bins=30,color="red")

    sns.violinplot(data=data[colum_name],ax=axes[1,1], palette="Set2",inner="quartile")

    sns.set_theme(style='ticks')

    plt.show()


def extremeValues(column_name):
    fig ,axes =plt.subplots(1,2, figsize=(12,8))
    sns.boxplot(data=data,y=column_name,ax=axes[0])
    sns.scatterplot(data=data[column_name],ax=axes[1])


    plt.show()
```

# Analyse 

### 1.histogramme 

Analyse de la variable job_description_length

L’analyse de l’histogramme représentant la variable job_description_length permet d’observer la distribution des longueurs des descriptions d’emploi dans le dataset.

1. Forme globale de la distribution

La distribution apparaît globalement uniforme sur l’intervalle étudié. Contrairement à une distribution normale en forme de cloche ou à une distribution asymétrique, les barres de l’histogramme présentent des hauteurs relativement similaires sur toute la plage des valeurs.

Cela indique qu’aucune longueur spécifique ne domine clairement dans les données.

On observe que les longueurs des descriptions sont réparties de manière homogène entre environ 500 et 2500 caractères. Il n’existe pas d’intervalle où les observations sont significativement plus nombreuses.

Cela suggère qu’il n’y a pas de standard unique adopté par les entreprises en matière de longueur de description.

3. Dispersion

La dispersion est relativement large, avec un minimum proche de 500 caractères et un maximum proche de 2500 caractères. Cette étendue importante indique une forte variabilité dans la manière dont les entreprises rédigent leurs offres d’emploi.

Certaines entreprises privilégient des descriptions courtes et synthétiques, tandis que d’autres fournissent des descriptions plus détaillées.

4. Présence de plusieurs pics

L’histogramme ne montre pas de présence de pics multiples. Il n’y a pas de regroupement distinct des valeurs qui indiquerait l’existence de sous-populations spécifiques.

Cela signifie que les données ne semblent pas segmentées en groupes distincts selon la longueur des descriptions.

5. Valeurs extrêmes

Aucune valeur extrême significative n’est visuellement identifiable sur le graphique. Les données semblent contenues dans un intervalle défini, sans observations anormalement élevées ou faibles.
