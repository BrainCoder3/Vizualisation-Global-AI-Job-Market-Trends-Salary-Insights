```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

La representation des diagrammes en barres pour les variables qualitatives.

```python
counts = data['job_title'].value_counts()
print(counts)

fig, ax = plt.subplots()

ax.bar(counts.index, counts.values, color='plum')
ax.set_ylabel('Frequence', color='darkblue')
ax.tick_params(axis='y', labelcolor='darkblue')

ax.set_xticks(range(len(counts)))
ax.set_xticklabels(counts.index, rotation=90, ha='center')

plt.title('Diagramme en barres (job_title)')
plt.show()
```

🔹 job_title

✔ Catégorie dominante :
Le poste le plus fréquent représente la plus grande part des offres. Il constitue le métier principal du marché.

✔ Écarts entre catégories :
Un écart important entre la première barre et les suivantes indique une forte spécialisation.

✔ Déséquilibre :
Si une seule catégorie domine largement, le marché est concentré sur peu de métiers.

✔ Catégories marginales :
Les postes très peu représentés correspondent à des fonctions rares ou spécialisées.

Dans cette section on va se concentrer sur la representation des diagrammes de pareto pour les variables qualitatives.

```python
#job_title

counts = data['job_title'].value_counts()
print(counts)
fig, ax1 = plt.subplots()

# Diagramme à barres
ax1.bar(counts.index, counts.values, color='plum')
ax1.set_ylabel('Frequence', color='darkblue')
ax1.tick_params(axis='y', labelcolor='darkblue')

# Rotation verticale des labels de x-axis
ax1.set_xticks(range(len(counts)))
ax1.set_xticklabels(counts.index, rotation=90, ha='center')

# Graphique linéaire du pourcentage cumulé
cum_percentage = counts.cumsum() / counts.sum() * 100 #calcule du pourcentage cumulé
ax2 = ax1.twinx()  # y-axis 2éme
ax2.plot(counts.index, cum_percentage, color='mediumvioletred', marker='o', linestyle='-')
ax2.set_ylabel('Pourcentage cumulé (%)', color='mediumvioletred')
ax2.tick_params(axis='y', labelcolor='mediumvioletred')
ax2.set_ylim(0, 110)  #se présenter jusqu'à 100%
plt.axhline(y=80, color='red', linestyle='--', label='80% line')
plt.title('Diagramme de Pareto(job_title)')
plt.show()
```

Analyse et interprétation:
-Ce diagramme de Pareto représente les intitulés de poste mondiaux de 20 emplois différents.
-Parmi tous ça, on distingue les 3 barres les plus hautes.: machine learning reasercher (808 avec frequence cumulé de 5,38%)
AI software engineer (784; 10,61%) et Autonomous Systems Engineer (777; 15,79%).
-Donc pour atteint 80% on a besion de plus de 20% de cause donc le principe de 80/20 ne s'applique pas car la courbe cumulative augmente lentement et n'atteint 80 % qu'au 16e ou 17e intitulé de poste, ce qui signifie que les fréquences sont très similaires dans toutes les catégories.
-Les autres intitulés de poste ont tous des fréquences similaires, entre 700 et 760, ce qui signifie qu'aucun intitulé de poste ne domine les autres.
-Étant donné que tous les intitulés de poste apparaissent à une fréquence quasi identique, il est difficile d'en privilégier un par rapport à un autre. Cela suggère que le marché de l'emploi est bien réparti entre toutes les spécialisations en IA.

La representation des diagrammes de circulaire pour les variables qualitatives.

```python
#création d'une fonction qui va construire nos diagrammes circulaires
def diagramme_circulaire(dataframe, colonne):
    # Calculer les fréquences
    counts = dataframe[colonne].value_counts()
    
    # Création du diagramme circulaire
    plt.figure(figsize=(6,6))
    plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=90)
    
    plt.title(f"Diagramme circulaire de {colonne}")
    plt.show()
```

```python
#job_title
diagramme_circulaire(data, "job_title")
```

Analyse et interprétation:
-Proportions relatives :
Chaque secteur du cercle représente la part de chaque poste dans l’ensemble des offres.
Plus le secteur est large, plus le poste est fréquent.
-Part dominante :
Le poste ayant la plus grande portion du cercle correspond au métier principal du marché.
